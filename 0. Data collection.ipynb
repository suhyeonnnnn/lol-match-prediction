데이터 열람 경로
- 리그(챌린저, 그랜드마스터, 마스터) 별 소환사 정보

- 닉네임으로 puuid 알아내기
- https://developer.riotgames.com/apis#summoner-v4
- 2번째 GET에서 닉네임 입력하면 puuid 추출 가능
	
- Puuid로 매치ID 알아내기
- https://developer.riotgames.com/apis#match-v5
- 1번째 GET에서 puuid 입력하면 최근 20게임 매치ID 추출 가능

- https://developer.riotgames.com/apis#match-v5/GET_getMatch
- 2번째 GET에서 매치ID 입력하면 한 게임의 정보 추출 가능
 
- Example
- https://asia.api.riotgames.com/lol/match/v5/matches/KR_5559807554?api_key=RGAPI-451c6e53-0a22-4ffa-87cc-fe76c45f47f9

#### CHALLENGER

```python
import requests
import pandas as pd
import requests
import json 
from tqdm import tqdm
import time
```

```python
challenger = 'https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=RGAPI-bcc3c0ab-d42a-4d19-851d-77b812182a79'
r = requests.get(challenger)
league_df = pd.DataFrame(r.json())
league_df.reset_index(inplace=True)
print(league_df)
```

```python
league_entries_df = pd.DataFrame(dict(league_df['entries'])).T #dict구조로 되어 있는 entries컬럼 풀어주기
league_df = pd.concat([league_df, league_entries_df], axis=1) #열끼리 결합

league_df = league_df.drop(['index', 'queue', 'name', 'leagueId', 'entries'], axis=1)
league_df.info()
league_df.to_csv('challenger.csv',index=False)#중간저장
```

## puuid 수집

```python
summonerNames = list(tier['summonerName'])
summonerNames
```

```python
api_key = 'RGAPI-4bb9e2ab-cdfd-4bd0-8404-8eec1954c49e'
puuid_urls = []

for summonerName in summonerNames:
    puuid_urls.append('https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/'+summonerName+'?api_key=' + api_key)
```

```python
len(puuid_urls)
```

```python
#데이터 프레임 생성
result = {
    "summonerId": [], #merge를 위해 id -> summonerName으로 변경
    "accountId": [],
    "puuid": [],
    "summonerName": [], #merge를 위해 name -> summonerName으로 변경
    "profileIconId": [],
    "revisionDate": [],
    "summonerLevel": []
}
```

```python
for puuid_url in tqdm(puuid_urls):
    r = requests.get(puuid_url)
    if r.status_code == 200: #없어진 summoner 제거
        game_data = json.loads(r.text)
        result["summonerId"].append(game_data["id"])
        result["accountId"].append(game_data["accountId"])
        result["puuid"].append(game_data["puuid"])
        result["summonerName"].append(game_data["name"])
        result["profileIconId"].append(game_data["profileIconId"])
        result["revisionDate"].append(game_data["revisionDate"])
        result["summonerLevel"].append(game_data["summonerLevel"])
        
    else:
        print("Error code")
```

```python
#데이터 프레임 변환 후 export
puuid = pd.DataFrame(result)
puuid.head()
```

```python
puuid.info()
```

```python
puuid.to_csv('puuid.csv', index=False)
```

## 경기 상세정보 수집

```python
import requests
import pandas as pd
import requests
import json 
from tqdm import tqdm
import time
```

```python
df = pd.read_csv('gamedetail_urls.csv')
```

```python
# 중복제거
df = df.drop_duplicates()
```

```python
#리스트로 만들기
df= df['0'].values.tolist()
```

```python
# api 키넣고 url 완성
gamedetail_urls = []
for i in df: #100개 게임 검색
    gamedetail_urls.append(i+'RGAPI-e1a73a49-3b4f-417b-86bb-5ff3d9097b3d')
```

```python
len(gamedetail_urls)
```

```python
# 환경적 제약(노트북 이동)으로 인해 데이터 프레임 끊어서 수집
```

```python
list_of_df =[]
for gamedetail_url in tqdm(gamedetail_urls):

    r = requests.get(gamedetail_url)
    dd = json.loads(r.text)
    
    try:
    # meta 풀기
        meta_data = pd.json_normalize(dd['metadata'], record_path='participants', meta = ['dataVersion','matchId'])
        meta_data.columns = ['puuid','dataVersion','matchId']

        # info - participants 풀기
        participants_data = pd.json_normalize(dd['info'], record_path=['participants'], 
                                              meta=['gameCreation','gameId','gameMode',
                                                   'gameName','gameType','gameVersion','mapId','platformId',
                                                   'queueId'])
        participants_data = participants_data.drop(['perks.styles','teamEarlySurrendered','wardsPlaced'], axis=1) 
                    #perk drop
                    #teams drop

        # 데이터 프레임 합치기
        df = pd.merge(participants_data,meta_data, how='left',on='puuid')

        list_of_df.append(df)
        
    except:
        print('key error')
    
    time.sleep(1) # 라이엇 게임즈 규정으로 인한 시간 지연
```

```python
len(list_of_df) #갯수 확인
```

```python
df_accum = pd.concat(list_of_df, axis=0) #데이터 프레임으로 변환
df_accum.to_csv('gamedetail.csv') # csv 파일로 내보내기
```
